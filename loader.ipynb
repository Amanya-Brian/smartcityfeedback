```python
from langchain_community.document_loaders import JSONLoader
import jsonlines, json
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
# loader = JSONLoader("test.jsonl", jq_schema = ".title, .content")

# text = loader.load()

# print(text)

# Open and read the JSONL file
with jsonlines.open('rwanda.jsonl') as reader:
    documents = [obj for obj in reader]

# documents
docs = []

# initialize text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=30)
for doc in documents:
    split = text_splitter.split_text(doc["content"])
    single_doc = text_splitter.create_documents(texts=split)
    docs.append(single_doc)

print(docs)
```
